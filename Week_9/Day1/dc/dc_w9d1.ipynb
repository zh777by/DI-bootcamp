```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('shopping_behavior_updated.csv')
df.head()
```

```python
# Age Distribution.

ages = df['Age']

# Plotting the histogram
plt.figure(figsize=(10, 6))
plt.hist(ages, bins=20, color='skyblue', edgecolor='black')
plt.title('Age Distribution of Customers')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.show()
```

```python
# Gender Analysis.

# Gender Distribution
plt.figure(figsize=(8, 6))
sns.countplot(x='Gender', data=df, palette='pastel', hue='Gender', dodge=False)
plt.title('Gender Distribution of Customers')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()

# Gender Preferences within Categories
plt.figure(figsize=(12, 8))
sns.countplot(x='Category', hue='Gender', data=df, palette='husl', dodge=False)
plt.title('Gender Preferences within Categories')
plt.xlabel('Category')
plt.ylabel('Count')
plt.legend(title='Gender')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
 # Item and Category Insights.

# Most Purchased Items
top_items = df['Item Purchased'].value_counts().head(10)
plt.figure(figsize=(12, 8))
sns.barplot(x=top_items.values, y=top_items.index, hue=top_items.index, palette='viridis', dodge=False)
plt.title('Top 10 Most Purchased Items')
plt.xlabel('Count')
plt.ylabel('Item')
plt.show()

# Most Purchased Categories
top_categories = df['Category'].value_counts().head(10)
plt.figure(figsize=(12, 8))
sns.barplot(x=top_categories.values, y=top_categories.index, hue=top_categories.index, palette='plasma', dodge=False)
plt.title('Top 10 Most Purchased Categories')
plt.xlabel('Count')
plt.ylabel('Category')
plt.show()

# Explore Revenue Contribution by Categories
plt.figure(figsize=(12, 8))
sns.barplot(x='Purchase Amount (USD)', y='Category', data=df, estimator=sum, errorbar=None, palette='magma', hue='Category', dodge=False)
plt.title('Revenue Contribution by Categories')
plt.xlabel('Total Purchase Amount (USD)')
plt.ylabel('Category')
plt.show()
```

```python
# Location Analysis

# Distribution of Customers across Locations
plt.figure(figsize=(16, 8))
sns.countplot(x='Location', data=df, hue='Location', palette='Set2', order=df['Location'].value_counts().index, dodge=False)
plt.title('Distribution of Customers across Locations')
plt.xlabel('Location')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.show()

# Explore Popular Products or Categories in Specific Regions
plt.figure(figsize=(16, 8))
sns.countplot(x='Location', hue='Category', data=df, palette='tab10', dodge=True, order=df['Location'].value_counts().index)
plt.title('Popular Products or Categories in Specific Regions')
plt.xlabel('Location')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Category', bbox_to_anchor=(1, 1))
plt.show()
```

```python
5.# Purchase Amount Analysis

# Distribution of Purchase Amounts
plt.figure(figsize=(12, 8))
sns.histplot(df['Purchase Amount (USD)'], bins=30, kde=True, color='skyblue')
plt.title('Distribution of Purchase Amounts')
plt.xlabel('Purchase Amount (USD)')
plt.ylabel('Frequency')
plt.show()

# Check for Outliers or High-Value Purchases
plt.figure(figsize=(12, 8))
sns.boxplot(x=df['Purchase Amount (USD)'], color='salmon')
plt.title('Boxplot of Purchase Amounts')
plt.xlabel('Purchase Amount (USD)')
plt.show()
```

```python
# Seasonal Trends

# Popularity of Items across Seasons
plt.figure(figsize=(14, 8))
sns.countplot(x='Season', hue='Item Purchased', data=df, palette='viridis', dodge=True)
plt.title('Popularity of Items across Seasons')
plt.xlabel('Season')
plt.ylabel('Count')
plt.legend(title='Item', bbox_to_anchor=(1, 1))
plt.show()

# Seasonal Preferences in Product Choices
plt.figure(figsize=(14, 8))
sns.countplot(x='Season', hue='Category', data=df, palette='muted', dodge=True)
plt.title('Seasonal Preferences in Product Choices')
plt.xlabel('Season')
plt.ylabel('Count')
plt.legend(title='Category', bbox_to_anchor=(1, 1))
plt.show()
```

```python
# Review Ratings.

# Distribution of Review Ratings
plt.figure(figsize=(10, 6))
sns.histplot(df['Review Rating'], bins=5, kde=True, color='green')
plt.title('Distribution of Review Ratings')
plt.xlabel('Review Rating')
plt.ylabel('Frequency')
plt.show()

# Explore Relationship between Review Ratings and Loyalty (Previous Purchases)
plt.figure(figsize=(10, 6))
sns.boxplot(x='Review Rating', y='Previous Purchases', data=df, color='orange')
plt.title('Review Ratings vs. Previous Purchases')
plt.xlabel('Review Rating')
plt.ylabel('Previous Purchases')
plt.show()
```

```python
# Subscription and Shipping Preferences

# Frequency of Purchases based on Subscription Status
plt.figure(figsize=(10, 6))
sns.countplot(x='Frequency of Purchases', hue='Subscription Status', data=df, palette='pastel')
plt.title('Frequency of Purchases based on Subscription Status')
plt.xlabel('Frequency of Purchases')
plt.ylabel('Count')
plt.legend(title='Subscription Status', bbox_to_anchor=(1, 1))
plt.show()

# Preferred Shipping Types and their Impact on Purchase Frequency
plt.figure(figsize=(12, 8))
sns.countplot(x='Shipping Type', hue='Frequency of Purchases', data=df, palette='muted', dodge=True)
plt.title('Preferred Shipping Types and Impact on Purchase Frequency')
plt.xlabel('Shipping Type')
plt.ylabel('Count')
plt.legend(title='Frequency of Purchases', bbox_to_anchor=(1, 1))
plt.show()
```

```python
# Discounts and Promo Codes

# Impact of Discounts on Purchase Frequency
plt.figure(figsize=(10, 6))
sns.countplot(x='Discount Applied', hue='Frequency of Purchases', data=df, palette='Set2', dodge=True)
plt.title('Impact of Discounts on Purchase Frequency')
plt.xlabel('Discount Applied')
plt.ylabel('Count')
plt.legend(title='Frequency of Purchases', bbox_to_anchor=(1, 1))
plt.show()

# Explore the Proportion of Purchases with Discounts
discount_counts = df['Discount Applied'].value_counts()
plt.figure(figsize=(8, 6))
plt.pie(discount_counts, labels=discount_counts.index, autopct='%1.1f%%', colors=['lightblue', 'lightcoral'])
plt.title('Proportion of Purchases with Discounts')
plt.show()
```

```python
# Payment Method Analysis

# Preferred Payment Methods
plt.figure(figsize=(10, 6))
sns.countplot(x='Payment Method', hue='Payment Method', data=df, palette='Blues', dodge=False)
plt.title('Preferred Payment Methods')
plt.xlabel('Payment Method')
plt.ylabel('Count')
plt.show()

# Analyze Purchase Amounts by Payment Method
plt.figure(figsize=(12, 8))
sns.boxplot(x='Payment Method', y='Purchase Amount (USD)', hue='Payment Method', data=df, palette='mako', dodge=False)
plt.title('Purchase Amounts by Payment Method')
plt.xlabel('Payment Method')
plt.ylabel('Purchase Amount (USD)')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Frequency of Purchases

# Examine the Frequency of Purchases
plt.figure(figsize=(12, 8))
sns.countplot(x='Frequency of Purchases', hue='Frequency of Purchases', data=df, palette='rocket', dodge=False)
plt.title('Frequency of Purchases')
plt.xlabel('Frequency of Purchases')
plt.ylabel('Count')
plt.show()

# Determine Time Intervals for Purchases
plt.figure(figsize=(12, 8))
sns.countplot(x='Frequency of Purchases', hue='Shipping Type', data=df, palette='viridis', dodge=True)
plt.title('Time Intervals for Purchases by Shipping Type')
plt.xlabel('Frequency of Purchases')
plt.ylabel('Count')
plt.legend(title='Shipping Type', bbox_to_anchor=(1, 1))
plt.show()
```
