```python
import pandas as pd
import plotnine as p9
import plotly.express as px
import matplotlib.pyplot as plt

df = pd.read_csv('2016.csv')
df
```

```python
df.dtypes
```

```python
df.isnull().sum()
```

```python
df.isna()
```

```python
df_sorted = df.sort_values(by='Happiness Score', ascending=False)

plt.figure(figsize=(15, 20))
plt.bar(df_sorted['Happiness Score'], df_sorted['Country'], color='skyblue')
plt.xlabel('Country')
plt.ylabel('Happiness Score')
plt.title('Happiness Scores Across Countries')
plt.xticks(rotation=90, ha='right')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
from plotnine import ggplot, aes, geom_point, theme_bw, labs

scatter_plot = ggplot(df, aes(x='Economy (GDP per Capita)', y='Happiness Score', color='Region', label='Country')) + \
               geom_point() + \
               labs(title='Scatter Plot: GDP per Capita vs Happiness Score',
                    x='GDP per Capita', y='Happiness Score', color='Region') + \
               theme_bw()

print(scatter_plot)
```

```python
correlation_matrix = df.corr()

fig = px.imshow(correlation_matrix, labels=dict(color='Correlation'),
                x=correlation_matrix.columns, y=correlation_matrix.columns,
                title='Correlation Heatmap')

fig.update_layout(
    width=900,
    height=900,
)

fig.show()
```

```python
fig = px.scatter(df,
                 x='Health (Life Expectancy)',
                 y='Freedom',
                 size='Happiness Score',
                 color='Region',
                 hover_name='Country',
                 title=('Interactive Scatter Plot of Health, Freedom, and Happiness'))

fig.show()
```

```python
constant_bubble_size = 10

fig = px.scatter(df,
                 x='Family',
                 y='Generosity',
                 size=[constant_bubble_size]*len(df),
                 color='Happiness Score',
                 hover_name='Country',
                 title='Bubble Chart: Social Support, Generosity, and Happiness')

fig.update_layout(xaxis_title='Social Support',
                  yaxis_title='Generosity')

fig.show()
```

```python
fig = px.scatter_3d(df,
                    x='Family',
                    y='Generosity',
                    z='Health (Life Expectancy)',
                    size='Happiness Score',
                    color='Freedom',
                    hover_name='Country',
                    title='Multi-Variable Analysis of Happiness',
                    labels={'Social Support': 'Social Support',
                            'Generosity': 'Generosity',
                            'Health': 'Health',
                            'Happiness Score': 'Happiness Score',
                            'Freedom': 'Freedom'})

fig.update_layout(scene=dict(xaxis_title='Social Support',
                             yaxis_title='Generosity',
                             zaxis_title='Health'))

fig.show()
```

```python
df_2019 = pd.read_csv('2019.csv')
```

```python
fig = px.line(df, x='Country or region', y='Freedom to make life choices',
              title='Happiness Score Comparison - 2016 vs 2019',
              labels={'Happiness Score': 'Happiness Score'},
              line_shape='linear')

fig.add_trace(px.line(df_2019, x='Country or region', y='Freedom to make life choices').data[0])

fig.update_layout(xaxis_title='Country', yaxis_title='Happiness Score')

fig.show()
```
