```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import numpy as np

df = pd.read_csv('Bicycle_Thefts_Open_Data.csv')
df
```

```python
df.dtypes
```

```python
df.isnull().sum()
```

```python
df['BIKE_MAKE'] = df['BIKE_MAKE'].replace('NA', np.nan)
df['BIKE_MODEL'] = df['BIKE_MODEL'].replace('NA', np.nan)
df['BIKE_COLOUR'] = df['BIKE_COLOUR'].replace('NA', np.nan)
```

```python
df.dropna(inplace=True)
```

```python
df.isna()
```

```python
# These visualizations identify patterns and trends in bicycle thefts over different time intervals.

df['OCC_DATE'] = pd.to_datetime(df['OCC_DATE'])

# Monthly Analysis
plt.figure(figsize=(12, 6))
monthly_counts = df.groupby(df['OCC_DATE'].dt.strftime('%B')).size()
monthly_counts.plot(kind='bar', xlabel='Month', ylabel='Number of Thefts', title='Monthly Bicycle Theft Trends')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Day of the Week Analysis
day_of_week_counts = df['OCC_DOW'].value_counts()
day_of_week_counts = day_of_week_counts.reindex(['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])
day_of_week_counts.plot(kind='bar', xlabel='Day of the Week', ylabel='Number of Thefts', title='Day of the Week Bicycle Theft Trends')
plt.show()

# Hourly Analysis
hourly_counts = df['OCC_HOUR'].value_counts().sort_index()
hourly_counts.plot(kind='bar', xlabel='Hour of the Day', ylabel='Number of Thefts', title='Hourly Bicycle Theft Trends')
plt.show()
```

```python
# Checking if there are specific time patterns when thefts occur.

stolen_bikes = df[df['PRIMARY_OFFENCE'].str.contains('THEFT', case=False, na=False)].copy()
stolen_bikes['OCC_DATE'] = pd.to_datetime(stolen_bikes['OCC_DATE'])

stolen_bikes.loc[:, 'Day_of_Week'] = stolen_bikes['OCC_DATE'].dt.day_name()
stolen_bikes.loc[:, 'Month'] = stolen_bikes['OCC_DATE'].dt.month_name()
stolen_bikes.loc[:, 'Hour'] = stolen_bikes['OCC_DATE'].dt.hour

# Plotting day of the week
plt.figure(figsize=(12, 6))
sns.countplot(x='Day_of_Week', data=stolen_bikes, order=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])
plt.title('Bicycle Thefts by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Number of Thefts')
plt.show()

# Plotting month
plt.figure(figsize=(12, 6))
sns.countplot(x='Month', data=stolen_bikes, order=['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December'])
plt.title('Bicycle Thefts by Month')
plt.xlabel('Month')
plt.ylabel('Number of Thefts')
plt.show()

# Plotting hour of the day
plt.figure(figsize=(12, 6))
sns.countplot(x='Hour', data=stolen_bikes)
plt.title('Bicycle Thefts by Hour of the Day')
plt.xlabel('Hour of the Day')
plt.ylabel('Number of Thefts')
plt.show()
```

```python
# The characteristics of stolen bikes, particularly focusing on the brand, model, and color

stolen_bikes = df[df['PRIMARY_OFFENCE'].str.contains('THEFT', case=False, na=False)]

# Count and visualize the top 10 bike makes
top_n = 10
top_bike_makes = stolen_bikes['BIKE_MAKE'].value_counts().nlargest(top_n)
plt.figure(figsize=(12, 6))
sns.barplot(x=top_bike_makes.values, y=top_bike_makes.index, palette='viridis')
plt.title(f'Top {top_n} Stolen Bike Makes')
plt.xlabel('Number of Stolen Bikes')
plt.ylabel('Bike Make')
plt.show()

# Count and visualize the top 10 bike models
top_bike_models = stolen_bikes['BIKE_MODEL'].value_counts().nlargest(top_n)
plt.figure(figsize=(12, 6))
sns.barplot(x=top_bike_models.values, y=top_bike_models.index, palette='viridis')
plt.title(f'Top {top_n} Stolen Bike Models')
plt.xlabel('Number of Stolen Bikes')
plt.ylabel('Bike Model')
plt.show()

# Count and visualize the top 10 bike colors
top_bike_colors = stolen_bikes['BIKE_COLOUR'].value_counts().nlargest(top_n)
plt.figure(figsize=(12, 6))
sns.barplot(x=top_bike_colors.values, y=top_bike_colors.index, palette='viridis')
plt.title(f'Top {top_n} Stolen Bike Colors')
plt.xlabel('Number of Stolen Bikes')
plt.ylabel('Bike Color')
plt.show()
```

Identifying potential predictive factors for bicycle thefts involves exploring various variables beyond the ones provided in the dataset.  Factors like weather, holidays, and other external elements could contribute to predicting bicycle thefts. Here are some additional factors you may want to consider:

Weather Conditions:

Temperature: Warmer weather might attract more people to outdoor activities, increasing the likelihood of theft.
Precipitation: Rain or snow might impact the frequency of bicycle usage and theft.
Time-related Factors:

Day of the Week: Weekends or specific weekdays may have higher theft rates.
Time of Day: Daytime or nighttime patterns may differ.
Holidays and Events:

Public Holidays: Theft rates may vary during holidays when people are more or less likely to use bicycles.
Special Events: Local events or festivals may influence theft rates.
Economic Factors:

Economic Conditions: Economic downturns or upturns may impact theft rates.
Unemployment Rates: Higher unemployment might correlate with increased theft.
Law Enforcement and Security Measures:

Policing Strategies: Changes in policing strategies or presence may affect theft rates.
Security Measures: Implementation of bike racks, surveillance, or other security measures.
Community and Social Factors:

Population Density: Higher population density areas might have more thefts.
Community Engagement: Active community programs may deter theft.
Previous Incidents:

Historical Data: Previous thefts in an area might indicate a pattern.
Bike Characteristics:

Expensive Bikes: High-value bikes may be more attractive to thieves.
Visibility: Highly visible or customized bikes might be targeted more often.
