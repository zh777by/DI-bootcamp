```python
from sklearn.preprocessing import LabelEncoder

binary_features = ['blue', 'dual_sim', 'four_g', 'three_g', 'touch_screen', 'wifi']

label_encoder = LabelEncoder()
for feature in binary_features:
    df[feature] = label_encoder.fit_transform(df[feature])
```

```python
from scipy.stats import skew, kurtosis

central_tendency = df.aggregate(['mean', 'median', lambda x: x.mode().iloc[0]], numeric_only=True)
print("Central Tendency Measures:")
print(central_tendency)

variability = pd.DataFrame({
    'Range': df.max() - df.min(),
    'Variance': df.var(),
    'Standard Deviation': df.std()
})
print("\nVariability Measures:")
print(variability)

distribution_shape = pd.DataFrame({
    'Skewness': df.apply(skew),
    'Kurtosis': df.apply(kurtosis)
})
print("\nDistribution Shape Measures:")
print(distribution_shape)
```

```python
import scipy.stats as stats

price_range_0_data = df[df['price_range'] == 0]['ram']
price_range_1_data = df[df['price_range'] == 1]['ram']

t_statistic, p_value = stats.ttest_ind(price_range_0_data, price_range_1_data)

alpha = 0.05
print("t-statistic:", t_statistic)
print("p-value:", p_value)

if p_value < alpha:
    print("Reject the null hypothesis. There is a significant difference between groups.")
else:
    print("Fail to reject the null hypothesis. There is no significant difference between groups.")
```

```python
import scipy.stats as stats

target_variable = 'price_range'

correlations, p_values = [], []

for feature in df.columns:
    if feature != target_variable:
        correlation, p_value = stats.pearsonr(df[feature], df[target_variable])
        correlations.append(correlation)
        p_values.append(p_value)

correlation_results = pd.DataFrame({
    'Feature': df.columns[df.columns != target_variable],
    'Pearson_Correlation': correlations,
    'P_Value': p_values
})

print(correlation_results)
```

```python
from scipy.stats import kendalltau

correlations_tau, p_values_tau = [], []

for feature in df.columns:
    if feature != target_variable:
        correlation_tau, p_value_tau = kendalltau(df[feature], df[target_variable])
        correlations_tau.append(correlation_tau)
        p_values_tau.append(p_value_tau)

correlation_results_tau = pd.DataFrame({
    'Feature': df.columns[df.columns != target_variable],
    'Kendall_Correlation': correlations_tau,
    'P_Value_Kendall': p_values_tau
})

print(correlation_results_tau)
```

```python
from scipy.stats import pointbiserialr

binary_variable = 'wifi'

correlation_pb, p_value_pb = pointbiserialr(df[binary_variable], df[target_variable])

print(f"Point-Biserial Correlation for {binary_variable} and {target_variable}: {correlation_pb}")
print(f"P-Value: {p_value_pb}")
```

```python
from scipy.stats import chi2_contingency

categorical_variable = 'three_g'

contingency_table = pd.crosstab(df[categorical_variable], df[target_variable])

chi2_stat, p_value_chi2, _, _ = chi2_contingency(contingency_table)

print(f"Chi-square statistic: {chi2_stat}")
print(f"P-Value: {p_value_chi2}")
```
