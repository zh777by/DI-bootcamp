```python
import pandas as pd
```

```python
df = pd.read_csv('globalterrorismdb_0718dist.csv', encoding='latin1')
df.head(5)
```

```python
missing_values = df.isnull().sum()
print(missing_values[missing_values > 0])
```

```python
print(df.dtypes)
```

```python
# Display unique values for each column
for column in df.columns:
    print(f"\n{column} unique values:\n{df[column].unique()}")

print(df.describe())
```

```python
df_cleaned = df.dropna()
```

```python
from sklearn.preprocessing import MinMaxScaler

numerical_columns = df.select_dtypes(include=['float64', 'int64']).columns

scaler = MinMaxScaler()

df[numerical_columns] = scaler.fit_transform(df[numerical_columns])

print("DataFrame after Min-Max Scaling:")
print(df)
```

```python
from sklearn.preprocessing import StandardScaler

numerical_columns = df.select_dtypes(include=['float64', 'int64']).columns

scaler = StandardScaler()

df[numerical_columns] = scaler.fit_transform(df[numerical_columns])

print("DataFrame after Standardization:")
print(df)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Visualize the distribution of attacks over the years
plt.figure(figsize=(12, 6))
sns.countplot(x='iyear', data=df, palette='viridis')
plt.title('Number of Terrorist Attacks Over the Years')
plt.xlabel('Year')
plt.ylabel('Number of Attacks')
plt.show()

# Visualize the distribution of attacks by region
plt.figure(figsize=(14, 8))
sns.countplot(x='region_txt', data=df, palette='Set2', order=df['region_txt'].value_counts().index)
plt.title('Number of Terrorist Attacks by Region')
plt.xlabel('Region')
plt.ylabel('Number of Attacks')
plt.xticks(rotation=45, ha='right')
plt.show()

# Visualize the correlation between numerical variables
correlation_matrix = df[['nkill', 'nwound', 'propvalue']].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix of Numeric Variables')
plt.show()

# Boxplot to visualize the distribution of casualties by attack type
plt.figure(figsize=(16, 8))
sns.boxplot(x='attacktype1_txt', y='nkill', data=df, palette='husl')
plt.title('Distribution of Casualties by Attack Type')
plt.xlabel('Attack Type')
plt.ylabel('Number of Killings')
plt.xticks(rotation=45, ha='right')
plt.show()

# Scatter plot to explore the relationship between socio-economic variables
plt.figure(figsize=(12, 8))
sns.scatterplot(x='propvalue', y='nkill', data=df, hue='success', palette='coolwarm', alpha=0.7)
plt.title('Scatter Plot: Property Value vs. Number of Killings')
```
