```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler, StandardScaler


file_path = 'adult.data'
column_names = ['age', 'workclass', 'fnlwgt', 'education', 'education-num', 'marital-status', 'occupation', 'relationship', 'race', 'sex', 'capital-gain', 'capital-loss', 'hours-per-week', 'native-country', 'income']
df = pd.read_csv('/Users/macbook/Desktop/DI_bootcamp/Week_8/Day3/dc/adult/adult.data', names=column_names, na_values=' ?')
df.head()
```

```python
# Identify numerical columns
numerical_columns = ['age', 'fnlwgt', 'education-num', 'capital-gain', 'capital-loss', 'hours-per-week']

# Extract numerical data
numerical_data = df[numerical_columns]

# Min-Max scaling
min_max_scaler = MinMaxScaler()
min_max_scaled_data = min_max_scaler.fit_transform(numerical_data)
min_max_scaled_df = pd.DataFrame(min_max_scaled_data, columns=numerical_columns)

# Z-score normalization
z_score_scaler = StandardScaler()
z_score_normalized_data = z_score_scaler.fit_transform(numerical_data)
z_score_normalized_df = pd.DataFrame(z_score_normalized_data, columns=numerical_columns)

print("\nMin-Max Scaled Dataset:")
print(min_max_scaled_df.head())

print("\nZ-score Normalized Dataset:")
print(z_score_normalized_df.head())
```

```python
# Z-score method for outliers
z_scores = np.abs((numerical_data - numerical_data.mean()) / numerical_data.std())
outliers_zscore = (z_scores > 3).any(axis=1)

# IQR method for outliers
Q1 = numerical_data.quantile(0.25)
Q3 = numerical_data.quantile(0.75)
IQR = Q3 - Q1
outliers_iqr = ((numerical_data < (Q1 - 1.5 * IQR)) | (numerical_data > (Q3 + 1.5 * IQR))).any(axis=1)

print("Outliers using Z-score method:")
print(df[outliers_zscore])

print("\nOutliers using IQR method:")
print(df[outliers_iqr])
```

```python
# Define the winsorization threshold (e.g., 99th percentile)
winsorizing_threshold = 0.99

# Apply winsorizing to the dataset
data_winsorized = df.copy()

for column in numerical_columns:
    max_threshold = data_winsorized[column].quantile(winsorizing_threshold)
    data_winsorized[column] = np.where(data_winsorized[column] > max_threshold, max_threshold, data_winsorized[column])

print("Winsorized Dataset:")
print(data_winsorized)
```
