```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
from scipy.stats import norm
import pandas as pd
```

```python
data = [12, 15, 13, 12, 18, 20, 22, 21]

mean = np.mean(data)
median = np.median(data)
variance = np.var(data, ddof=1)
std_deviation = np.std(data, ddof=1)

print("Mean:", mean)
print("Median:", median)
print("Variance:", variance)
print("Standard Deviation:", std_deviation)
```

```python
mean_value = 50
std_deviation_value = 10

data = np.random.normal(loc=mean_value, scale=std_deviation_value, size=1000)

plt.hist(data, bins=30, density=True, alpha=0.7, color='blue')

xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mean_value, std_deviation_value)
plt.plot(x, p, 'k', linewidth=2)

plt.title('Normal Distribution: Mean = 50, Std Dev = 10')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.show()
```
