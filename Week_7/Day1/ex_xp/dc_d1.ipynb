```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
np.random.seed(42)  
years = np.arange(2000, 2022)
num_years = len(years)
num_months = 12
temperature_data = np.random.uniform(low=-10, high=30, size=(num_years, num_months))

average_yearly_temperature = np.mean(temperature_data, axis=1)

columns = [f'Month_{i+1}' for i in range(num_months)]
temperature_df = pd.DataFrame(data=temperature_data, index=years, columns=columns)
average_yearly_temperature = temperature_df.mean(axis=1)

plt.figure(figsize=(10, 6))
plt.plot(years, average_yearly_temperature, marker='o', linestyle='-', color='b')
plt.title('Global Temperature Trends (2000-2021)')
plt.xlabel('Year')
plt.ylabel('Average Yearly Temperature (°C)')
plt.grid(True)
plt.show()
```

```python
np.random.seed(42)

num_cities = 10
num_months = 12

temperature_data = np.random.uniform(low=-5, high=35, size=(num_cities, num_months))

city_labels = [f'City_{i+1}' for i in range(num_cities)]
month_labels = [f'Month_{i+1}' for i in range(num_months)]

temperature_df = pd.DataFrame(data=temperature_data, index=city_labels, columns=month_labels)

annual_avg_temperature = temperature_df.mean(axis=1)

print("Synthetic Temperature DataFrame:")
print(temperature_df)
print("\nAnnual Average Temperature for Each City:")
print(annual_avg_temperature)
```
