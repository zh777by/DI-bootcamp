```python
#exo 2


import urllib.request

# URL of the website's robots.txt file
url = 'https://en.wikipedia.org/robots.txt'

try:
    # Open the URL and read the content
    with urllib.request.urlopen(url) as response:
        content = response.read()

    # Decode the content to a readable format
    content = content.decode('utf-8')

    # Print the content of robots.txt
    print(content)

except Exception as e:
    print(f"An error occurred: {e}")

```

```python
## exo 3

import requests
from bs4 import BeautifulSoup

# URL of the Wikipedia main page
url = 'https://en.wikipedia.org/wiki/Main_Page'

# Fetch the content of the URL
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Parse the content using BeautifulSoup
    soup = BeautifulSoup(response.content, 'html.parser')

    # Find all header tags (h1, h2, h3, h4, h5, h6)
    headers = soup.find_all(['h1', 'h2', 'h3', 'h4', 'h5', 'h6'])

    # Print each header tag
    for header in headers:
        print(header.get_text().strip())
else:
    print(f"Failed to retrieve the webpage. Status code: {response.status_code}")
```

```python
## exo 4

import requests
from bs4 import BeautifulSoup

# URL of the web page you want to check
url = 'https://en.wikipedia.org/wiki/Main_Page'  # You can replace this with any URL

# Fetch the content of the URL
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Parse the content using BeautifulSoup
    soup = BeautifulSoup(response.content, 'html.parser')

    # Find the title tag
    title = soup.find('title')

    # Check if title tag is found and print the result
    if title:
        print(f"Page has a title: {title.get_text()}")
    else:
        print("Page does not have a title tag.")
else:
    print(f"Failed to retrieve the webpage. Status code: {response.status_code}")
```
