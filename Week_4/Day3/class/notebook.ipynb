```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots()  # Create a figure containing a single axes
x = [1, 2, 3, 4, 5]  # Prices
y = [5, 4, 3, 2, 1]  # Demand
ax.set(title='Product Demand vs Price', xlabel='Price', ylabel='Demand')
ax.plot(x, y)
fig.savefig('my_new_figure_1.png') 
```

```python
x = [1, 2, 3, 4, 5]  # Age
y = [75, 80, 88, 95, 105]  # Height in cm
plt.scatter(x, y)

ax.set(title="Children's Height vs Age", xlabel='Age (years)', ylabel='Height (cm)')

plt.show()
fig.savefig('my_new_figure_2.png')
```

```python
import numpy as np

x = np.linspace(-10, 10, 1000)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, color='tab:blue', label='sin(x)')
plt.scatter(x, y2, marker='o', color='tab:blue', label='cos(x)')

plt.legend()
plt.show()
```
